```python
from sklearn.feature_extraction.text import CountVectorizer
import os
import nltk
from urllib.parse import urlsplit
import shutil

file_store = 'D:/Khai_Pha_Web/BTN_C4/Data_Preprocessing'
text_pre = []
for file in os.listdir(file_store):
    with open(os.path.join(file_store, file), 'r', encoding='utf-8') as f:
        text = f.read()
        text_pre.append(text)

# print(text_pre)
# Xay dung vector BOW
vect = CountVectorizer()
X = vect.fit_transform(text_pre)

# Xay dung tu dien
dictionary=list(vect.get_feature_names_out())

print("Words in dictionary: ", dictionary)
print("Vector Bag-of-Word: \n", X.toarray())
```
