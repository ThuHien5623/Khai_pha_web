```python
import os
import nltk
from urllib.parse import urlsplit
import shutil
from wordcloud import WordCloud
import matplotlib.pyplot as plt

file_store = 'D:/Khai_Pha_Web/BTN_C4/Data_Preprocessing'

text_pre = ''
for file in os.listdir(file_store):
    with open(os.path.join(file_store, file), 'r', encoding='utf-8') as f:
        text = f.read()
        text_pre += text
        
# Convert the text to a dictionary of word frequencies / Chuyn i vn bn thnh t in tn s t
word_counts = nltk.FreqDist(text_pre.split())
# Generate the word cloud / To m my t
wcloud = WordCloud().generate_from_frequencies(word_counts)
# Plotting the wordcloud / V s  m my t
plt.imshow(wcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
